```python
train_data=train_data.drop(['acq_sub_chn'],axis=1)
test_data=test_data.drop(['acq_sub_chn'],axis=1)
test_data=test_data.drop(['acq_type_grp'],axis=1)
train_data=train_data.drop(['acq_type_grp'],axis=1)
```

```python
print(train_data.shape, test_data.shape)
```

```python
train_data=train_data.drop(['cnsumr_chrg_avg_credit_12m_amt'],axis=1)
test_data=test_data.drop(['cnsumr_chrg_avg_credit_12m_amt'],axis=1)
train_data=train_data.drop(['sow_tot_annual_ext_pmt_amt'],axis=1)
test_data=test_data.drop(['sow_tot_annual_ext_pmt_amt'],axis=1)
train_data=train_data.drop(['sow_tot_revol_bal_amt'],axis=1)
test_data=test_data.drop(['sow_tot_revol_bal_amt'],axis=1)
train_data=train_data.drop(['sow_tot_trans_bal_amt'],axis=1)
test_data=test_data.drop(['sow_tot_trans_bal_amt'],axis=1)
```

```python
train_data = train_data.merge(pd.get_dummies(train_data.fee_type_grp, drop_first=True, prefix='fee_type_grp'), left_index=True, right_index=True)
test_data = test_data.merge(pd.get_dummies(train_data.fee_type_grp, drop_first=True, prefix='fee_type_grp'), left_index=True, right_index=True)
```

```python
train_data= train_data.drop(['cust_id'],axis=1)
test_data=test_data.drop(['cust_id'],axis=1)
```
